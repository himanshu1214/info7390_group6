```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
import pandas as pd
```

```python
Q_2006 = pd.read_csv("./zipfile/sample_orig_2006.csv", header= None, index_col= None  )
Q_2006.drop(Q_2006.columns[25], axis= 1, inplace = True)
Q_2006.columns = ['CREDIT SCORE', 'FIRST PAYMENT DATE', 'FIRST TIME HOMEBUYER FLAG', 'MATURITY DATE', 'MSA', 
              'MORTGAGE INSURANCE PERCENTAGE', 'NUMBER OF UNITS', 'OCCUPANCY STATUS', 'ORIGINAL COMBINED LOAN-TO-VALUE',
             'ORIGINAL DEBT-TO-INCOME (DTI) RATIO', 'ORIGINAL UPB', 'ORIGINAL LOAN-TO-VALUE (LTV)', 
             'ORIGINAL INTEREST RATE', 'CHANNEL', 'PREPAYMENT PENALTY MORTGAGE (PPM) FLAG', 'PRODUCT TYPE', 
                    'PROPERTY STATE', 'PROPERTY TYPE', 'POSTAL CODE', 'LOAN SEQUENCE NUMBER', 'LOAN PURPOSE', 
                    'ORIGINAL LOAN TERM', 'NUMBER OF BORROWERS', 'SELLER NAME', 'SERVICER NAME']
```

```python
Q1_2006_f = Q_2006[Q_2006["LOAN SEQUENCE NUMBER"].str.contains("Q1")]
Q2_2006_f = Q_2006[Q_2006["LOAN SEQUENCE NUMBER"].str.contains("Q2")]

Q1_2006_csv = Q1_2006_f.to_csv("Q1_2006.csv")
Q2_2006_csv = Q2_2006_f.to_csv("Q2_2006.csv")
```

```python
Q1_2006 = h2o.import_file("Q1_2006.csv")
Q2_2006 =  h2o.import_file("Q2_2006.csv")
```

```python
x = Q1_2006.columns
y = "ORIGINAL INTEREST RATE"
x.remove(y)
aml = H2OAutoML(max_models = 30, max_runtime_secs=300, seed = 1)
aml.train(x= x, y = y, training_frame= Q2_2006, leaderboard_frame= Q2_2006)
```
