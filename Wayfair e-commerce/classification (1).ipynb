```python
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
import pandas as pd

data01 = pd.read_csv('final_output_data.csv')


 

#give the target variable, the outcome(energy consumption)
y = data01['final_ordered_total']

#give the predictor variables
x = data01.drop(['customer_id','final_ordered_total'], axis=1)
x = x.drop(x.columns[0],axis =1)
print(x.head())
x = pd.get_dummies(x).values




X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.75, test_size=0.25)


X_train.shape, X_test.shape, y_train.shape, y_test.shape
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

tpot = TPOTClassifier(generations=1,verbosity=2)


tpot = TPOTClassifier(verbosity=2, max_time_mins=200)
tpot.fit(X_train, y_train)
print(tpot.score(X_test, y_test))

tpot.export('tpot_changedata_pipeline.py')

```
