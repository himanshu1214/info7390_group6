```python
df_sel = df_merge_tab.sample(frac=1)
df_ran_sel = df_sel.head(100000)
df_sel.head()
```

```python
df_ran_sel.to_csv('new_redu_merge.csv')
```

```python
df_ran_sel.date=df_ran_sel.date.str.replace('.*16$','Monday')
df_ran_sel.date=df_ran_sel.date.str.replace('.*17$','Tuesday')
df_ran_sel.date=df_ran_sel.date.str.replace('.*18$','Wednesday')
df_ran_sel.date=df_ran_sel.date.str.replace('.*19$','Thursday')
df_ran_sel.date=df_ran_sel.date.str.replace('.*20$','Friday')
df_ran_sel.date=df_ran_sel.date.str.replace('.*21$','Saturaday')
df_ran_sel.date=df_ran_sel.date.str.replace('.*22$','Sunday')
df_ran_sel.head()
```

```python
for i in range(0,12):
    if (i <10 ):
        df_ran_sel.time=df_ran_sel.time.str.replace('^0'+str(i)+':\S+','morning')
    df_ran_sel.time=df_ran_sel.time.str.replace('^'+str(i)+':\S+','morning')
for i in range(12,18):
    df_ran_sel.time=df_ran_sel.time.str.replace('^'+str(i)+':\S+','afternoon')
for i in range(18,24):  
    df_ran_sel.time=df_ran_sel.time.str.replace('^'+str(i)+':\S+','evening')
```

```python
import h2o
from h2o.automl import H2OAutoML
import pandas as pd
import numpy as np
from h2o.automl import H2OAutoML

h2o.init()

train = h2o.import_file("final_output_data.csv")

test = h2o.import_file("data1.csv")

# Identify predictors and response
x = train.columns
y = "final_ordered_total"
x.remove(y)
x.remove('customer_id')





# For binary classification, response should be a factor
train["final_ordered_total"] = train["final_ordered_total"].asfactor()
test["final_ordered_total"] = test["final_ordered_total"].asfactor()

# Run AutoML for 20 base models (limited to 1 hour max runtime by default)
aml = H2OAutoML(max_models=20, seed=1)
aml.train(x=x, y=y, training_frame=train)

# View the AutoML Leaderboard
lb = aml.leaderboard
lb.head(rows=lb.nrows)  # Print all rows instead of default (10 rows)



# The leader model is stored here
aml.leader

# If you need to generate predictions on a test set, you can make
# predictions directly on the `"H2OAutoML"` object, or on the leader
# model object directly

#preds = aml.predict(test)
```
